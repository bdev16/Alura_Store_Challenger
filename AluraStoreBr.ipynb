#1. Análise do faturamento


```python
def calcula_faturamento_loja(df_loja: pd.DataFrame) -> float:
  '''
  Vai realizar o calculo de faturamento de uma loja, pegando como base o DataFrame que contem seus dados.
  É necessário enviar um tipo dataframe por conta das instruções de conversão contidadas na função

  Retorno: O retorno da função vai ser um float com o valor do faturamento da loja
  '''
  df_modificado_loja = df_loja[['Produto', 'Preço', 'Frete']]

  dct_loja = {
      "Produtos": [],
      "Qtd_produtos": [],
      "Preço": [],
      "Frete": []
  }

  for produto, preco, frete in zip(df_modificado_loja['Produto'], df_modificado_loja['Preço'], df_modificado_loja['Frete']):
    if produto not in dct_loja['Produtos']:
        dct_loja["Produtos"].append(produto)
        dct_loja["Qtd_produtos"].append(1)
        dct_loja["Preço"].append(preco)
        dct_loja["Frete"].append(frete)
    else:
        indexProduto = dct_loja["Produtos"].index(produto)
        dct_loja["Qtd_produtos"][indexProduto] += 1

  faturamento_total_loja = 0;

  for quantidade, preco, frete in zip(dct_loja['Qtd_produtos'], dct_loja['Preço'], dct_loja['Frete']):
      faturamento_total_loja += (preco * quantidade) + frete

  return faturamento_total_loja
```

```python
faturamento_loja = calcula_faturamento_loja(loja)
faturamento_loja2 = calcula_faturamento_loja(loja2)
faturamento_loja3 = calcula_faturamento_loja(loja3)
faturamento_loja4 = calcula_faturamento_loja(loja4)

faturamento_lojas = (('Loja 1', faturamento_loja), ('Loja 2', faturamento_loja2), ('Loja 3', faturamento_loja3), ('Loja 4', faturamento_loja4))

pd.DataFrame(faturamento_lojas, columns=['Lojas', 'Faturamento'])
```

# 3. Média de Avaliação das Lojas

```python
def calcula_media_avaliacao_loja(df_loja: pd.DataFrame) -> float:
  '''
  Vai realizar o calculo da média de avaliação de uma loja, pegando como base o DataFrame que contem seus dados.
  É necessário enviar um tipo dataframe por conta das instruções de conversão contidadas na função

  Retorno: O retorno da função vai ser um float com a média
  '''

  avaliacao = df_loja['Avaliação da compra']
  total_notas = 0

  for nota in avaliacao:
    total_notas += nota

  media_avaliacao = total_notas / len(avaliacao)

  return media_avaliacao
```

```python
media_avalicao_loja = calcula_media_avaliacao_loja(loja)
media_avalicao_loja2 = calcula_media_avaliacao_loja(loja2)
media_avalicao_loja3 = calcula_media_avaliacao_loja(loja3)
media_avalicao_loja4 = calcula_media_avaliacao_loja(loja4)

media_avaliacao_lojas = (('Loja 1', media_avalicao_loja), ('Loja 2', media_avalicao_loja2), ('Loja 3', media_avalicao_loja3), ('Loja 4', media_avalicao_loja4))

pd.DataFrame(media_avaliacao_lojas, columns=['Lojas', 'Média de Avaliação'])
```

# 4. Produtos Mais e Menos Vendidos

```python
def produtos_mais_vendidos(df_loja: pd.DataFrame) -> pd.DataFrame:

  produtos = df_loja['Produto']
  dct_produtos = {
      "Produtos": [],
      "Qtd_vendas_produtos": []
  }

  for produto in produtos:
    if produto not in dct_produtos['Produtos']:
        dct_produtos["Produtos"].append(produto)
        dct_produtos["Qtd_vendas_produtos"].append(1)
    else:
        indexProduto = dct_produtos["Produtos"].index(produto)
        dct_produtos["Qtd_vendas_produtos"][indexProduto] += 1

  df_produtos = pd.DataFrame(dct_produtos)

  media_vendas_produtos = df_produtos['Qtd_vendas_produtos'].mean()
  produtos_mais_vendidos = df_produtos[df_produtos['Qtd_vendas_produtos'] >= media_vendas_produtos]
  produtos_mais_vendidos = produtos_mais_vendidos.sort_values(by='Qtd_vendas_produtos', ascending=False)

  return produtos_mais_vendidos.head(10)

def produtos_menos_vendidos(df_loja: pd.DataFrame) -> pd.DataFrame:

  produtos = df_loja['Produto']
  dct_produtos = {
      "Produtos": [],
      "Qtd_vendas_produtos": []
  }

  for produto in produtos:
    if produto not in dct_produtos['Produtos']:
        dct_produtos["Produtos"].append(produto)
        dct_produtos["Qtd_vendas_produtos"].append(1)
    else:
        indexProduto = dct_produtos["Produtos"].index(produto)
        dct_produtos["Qtd_vendas_produtos"][indexProduto] += 1

  df_produtos = pd.DataFrame(dct_produtos)

  media_vendas_produtos = df_produtos['Qtd_vendas_produtos'].mean()
  produtos_menos_vendidos = df_produtos[df_produtos['Qtd_vendas_produtos'] <= media_vendas_produtos]
  produtos_menos_vendidos = produtos_menos_vendidos.sort_values(by='Qtd_vendas_produtos', ascending=True)

  return produtos_menos_vendidos.head(10)
```
